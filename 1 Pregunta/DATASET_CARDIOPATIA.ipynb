```python
media = datos["IMC"].mean()
mediana = datos["IMC"].median()
moda = datos["IMC"].mode()
print("IMC")
print("""
    Media: %d
    Mediana: %d
    Moda: %d
""" % (media,mediana,moda))

media = datos["Salud_Fisica"].mean()
mediana = datos["Salud_Fisica"].median()
moda = datos["Salud_Fisica"].mode()
print("SALUD FISICA")
print("""
    Media: %d
    Mediana: %d
    Moda: %d
""" % (media,mediana,moda))

media = datos["Salud_Mental"].mean()
mediana = datos["Salud_Mental"].median()
moda = datos["Salud_Mental"].mode()
print("SALUD MENTAL")
print("""
    Media: %d
    Mediana: %d
    Moda: %d
""" % (media,mediana,moda))

media = datos["TiempodeDormir"].mean()
mediana = datos["TiempodeDormir"].median()
moda = datos["TiempodeDormir"].mode()
print("TIEMPO DE DORMIR")
print("""
    Media: %d
    Mediana: %d
    Moda: %d
""" % (media,mediana,moda))
```

CALCULO DE LA MEDIA, MODA, CUARTIL Y PERCENTIL CON LIBRERIA

```python
import numpy as np
import statistics as stats
```

```python
# Calcular la media
mean = np.mean(datos["IMC"])
# Calcular la mediana
median = np.median(datos["IMC"])
# Calcular la moda
mode = stats.mode(datos["IMC"])
print("IMC")
print("Media:", mean)
print("Mediana:", median)
print("Moda:", mode)

# Calcular la media
mean = np.mean(datos["Salud_Fisica"])
# Calcular la mediana
median = np.median(datos["Salud_Fisica"])
# Calcular la moda
mode = stats.mode(datos["Salud_Fisica"])
print("")
print("Salud Fisica")
print("Media:", mean)
print("Mediana:", median)
print("Moda:", mode)

# Calcular la media
mean = np.mean(datos["Salud_Mental"])
# Calcular la mediana
median = np.median(datos["Salud_Mental"])
# Calcular la moda
mode = stats.mode(datos["Salud_Mental"])
print("")
print("Salud Mental")
print("Media:", mean)
print("Mediana:", median)
print("Moda:", mode)

# Calcular la media
mean = np.mean(datos["TiempodeDormir"])
# Calcular la mediana
median = np.median(datos["TiempodeDormir"])
# Calcular la moda
mode = stats.mode(datos["TiempodeDormir"])
print("")
print("Tiempo de Dormir")
print("Media:", mean)
print("Mediana:", median)
print("Moda:", mode)
```
